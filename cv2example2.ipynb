```python
ret, binary = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)
```

```python
binary,contours,hierarchy= cv2.findContours(binary,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
print(binary.shape)
```

```python
# 将图像的区域分别画出来
for i in range(len(contours)):
    newMask= np.zeros([x,y,3])
    cv2.drawContours(newMask, contours, i, (255, 255, 255), -1)
    newMask = newMask[:,:,1]
    cv2.HoughLinesP
    cv2.namedWindow(""+str(i)+"",i);
    cv2.resizeWindow("enhanced", 640, 480);
    cv2.imshow(""+str(i)+"", newMask)
#cv2.imwrite("44.png", image)
cv2.waitKey(0)
```

```python
for i in range(len(contours)):
    newMask= np.zeros([x,y,3])
    #print(newMask)
    cv2.drawContours(newMask, contours, i, (1, 1, 1), -1)
    #newMask = cv2.cvtColor(newMask, cv2.COLOR_BGR2GRAY)
    newMask = np.array(newMask[:,:,1], dtype=np.uint8)
    instance_masks.append(newMask)
    print(instance_masks)
```
