# 生成SDLcode训练和测试图

```python
from PIL import Image, ImageDraw
import cv2
from skimage import io
import numpy as np
import os,sys
from skimage import measure,draw 
```

```python
img=Image.open("IDRiD_49_HE.png")  #IDRiD_49_HE.png  IDRiD_49_HEorg.png

orgimg=Image.open("IDRiD_49_HEorg.png")
img_gray=np.array(img.convert("L")) 
# val_MA = 178      #制作mask时，微血管瘤MA用（0，255，255）标注,转化为"L"模式灰度图，值为178
# val_HR = 29       #制作mask时，出血HR用（0，0，255）标注,转化为"L"模式灰度图，值为29
# val_HE = 105      #制作mask时，硬渗HE用（255，0，255）标注,转化为"L"模式灰度图，值为105
# val_SE = 225      #制作mask时，软渗SE用（255，255，0）标注,转化为"L"模式灰度图，值为225
        
# [x,y]=img_gray.shape
```

```python
val_HE = 105  #制作mask时，硬渗用（255，0，255）标注,转化为"L"模式灰度图，值为105
[x,y]=img_gray.shape
newMask= np.zeros([x,y,3], dtype = np.uint8)
org_img = np.array(orgimg).astype(np.uint8)

if val_HE in img_gray:
    mask_arr = np.array(img_gray==val_HE).astype(np.uint8)
    #mask_arr = mask_arr*255
    #binary,contours,hierarchy= cv2.findContours(mask_arr,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
    contours = measure.find_contours(mask_arr, 0.5)
    _contours = []
    for c in contours:
        c[:, [0, 1]] = c[:, [1, 0]]
        _contours.append(np.around(np.expand_dims(c, 1)).astype(np.int))

    for i in range(len(contours)):
        cv2.drawContours(org_img, _contours, i, (255, 255, 255), -1)
    
    newMask[:,:,0] = org_img[:,:,2]
    newMask[:,:,1] = org_img[:,:,1]
    newMask[:,:,2] = org_img[:,:,0]
    cv2.imwrite('train49Label.png',newMask)
```
